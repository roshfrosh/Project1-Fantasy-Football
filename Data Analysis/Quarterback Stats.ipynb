```python
import pandas as pd
import numpy as np
```

```python
QB = "GroupbyRoles/QB.csv"
quarterback = pd.read_csv(QB)
```

```python
QBPassingAvgYards = quarterback.groupby("Player Name")["Passing Avg Yards"].mean()
QBPassBlitz = quarterback.groupby("Player Name")["Passing Blitzes"].mean()
QBPassAvgCompletionPer = quarterback.groupby("Player Name")["Passing Completion Pct"].mean()
QBPassAvgCompletion = quarterback.groupby("Player Name")["Passing Completions"].mean()
QBPassDefended = quarterback.groupby("Player Name")["Passing Defended Passes"].mean()
QBPassDropped = quarterback.groupby("Player Name")["Passing Dropped Passes"].mean()
QBPassYards = quarterback.groupby("Player Name")["Passing Gross Yards"].mean()
QBPassHurries = quarterback.groupby("Player Name")["Passing Hurries"].mean()
QBPassIntercepts = quarterback.groupby("Player Name")["Passing Interceptions"].mean()
QBPassKnockdowns = quarterback.groupby("Player Name")["Passing Knockdowns"].mean()
QBPassLongest = quarterback.groupby("Player Name")["Passing Longest"].mean()
QBPassLongestTD = quarterback.groupby("Player Name")["Passing Longest TD"].mean()
QBPassNetYards = quarterback.groupby("Player Name")["Passing Net Yds"].mean()
QBPassPocketTime = quarterback.groupby("Player Name")["Passing Pocket Time"].mean()
QBPassRating = quarterback.groupby("Player Name")["Passing Rating"].mean()
QBPassRedZoneAttempts = quarterback.groupby("Player Name")["Passing Redzone Attempts"].mean()
QBPassSackYards = quarterback.groupby("Player Name")["Passing Sack Yds"].mean()
QBPassSacks = quarterback.groupby("Player Name")["Passing Sacks"].mean()
QBPassSpikes = quarterback.groupby("Player Name")["Passing Spikes"].mean()
QBPassThrowAway = quarterback.groupby("Player Name")["Passing Throw Away"].mean()
QBPassTouchdowns = quarterback.groupby("Player Name")["Passing Touchdowns"].mean()
QBPassYards = quarterback.groupby("Player Name")["Passing Yards"].mean()



QBPassingAverages = pd.DataFrame({
    'Avg Passing Average Yards': QBPassingAvgYards,
    'Avg Passing Blitz': QBPassBlitz,
    'Avg Passing Completion Percent': QBPassAvgCompletionPer,
    'Avg Passing Completions': QBPassAvgCompletion, 
    'Avg Passing Defended Passes': QBPassDefended,
    'Avg Passing Dropped Passes': QBPassDropped,
    'Avg Passing Gross Yards': QBPassYards,
    'Avg Passing Hurries': QBPassHurries,
    'Avg Passing Interceptions': QBPassIntercepts,
    'Avg Passing Knockdowns': QBPassKnockdowns,
    'Avg Passing Longest': QBPassLongest,
    'Avg Passing Longest TD': QBPassLongestTD,
    'Avg Passing Net Yds': QBPassNetYards,
    'Avg Passing Pocket Time': QBPassPocketTime,
    'Avg Passing Rating': QBPassRating,
    'Avg Passing Redzone Attempts': QBPassRedZoneAttempts,
    'Avg Passing Sack Yds': QBPassSackYards,
    'Avg Passing Sacks': QBPassSacks,
    'Avg Passing Spikes': QBPassSpikes,
    'Avg Passing Throw Away': QBPassThrowAway,
    'Avg Passing Touchdowns': QBPassTouchdowns,
    'Avg Passing Yards': QBPassYards
})
QBPassingAverages.head()
```

```python
output_data_QBPassing = "GroupbyRoles/QBPassing.csv"
QBPassingAverages.to_csv (output_data_QBPassing, header=True)
```
