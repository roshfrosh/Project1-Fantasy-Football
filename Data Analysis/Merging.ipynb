```python
output_data_final = "DataKeys/merged_data_2012_2018.csv"
player_data_final
receiving_stats_final
fumble_data_final
penalty_stats_final
rush_stats_final
punt_stats_final
puntreturn_stats_final
passing_stats_final
kickreturn_stats_final
kickoff_stats_final
interception_stats_final
fieldgoals_stats_final
extrapoints_stats_final
defense_stats_final
conversion_stats_final

df1 = pd.merge(player_data_final, receiving_stats_final, how = 'left', on="Key")
df2 = pd.merge(df1, fumble_data_final, how = 'left', on="Key")
df3 = pd.merge(df2, penalty_stats_final, how = 'left', on="Key")
df4 = pd.merge(df3, rush_stats_final, how = 'left', on="Key")
df5 = pd.merge(df4, punt_stats_final, how = 'left', on="Key")
df6 = pd.merge(df5, puntreturn_stats_final, how = 'left', on="Key")
df7 = pd.merge(df6, passing_stats_final, how = 'left', on="Key")
df8 = pd.merge(df7, kickreturn_stats_final, how = 'left', on="Key")
df9 = pd.merge(df8, kickoff_stats_final, how = 'left', on="Key")
df10 = pd.merge(df9, interception_stats_final, how = 'left', on="Key")
df11 = pd.merge(df10, fieldgoals_stats_final, how = 'left', on="Key")
df12 = pd.merge(df11, extrapoints_stats_final, how = 'left', on="Key")
df13 = pd.merge(df12, defense_stats_final, how = 'left', on="Key")
merged_data_2012_2018 = pd.merge(df13, conversion_stats_final, how = 'left', on="Key")
merged_data_2012_2018.to_csv (output_data_final, index = None, header=True)
merged_data_2012_2018.head()
```
