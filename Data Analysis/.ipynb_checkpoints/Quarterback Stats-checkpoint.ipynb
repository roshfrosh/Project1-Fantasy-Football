```python
import pandas as pd
import numpy as np
```

```python
QB = "GroupbyRoles/QB.csv"
quarterback = pd.read_csv(QB)
scoring = "FantasyStats/statsfactor.csv"
scores = pd.read_csv(scoring)
```

```python
quarterback["Passing Yards Score"] = quarterback["Passing Yards"] * scores.iloc[0, 0]
quarterback["TD Pass Score"] = quarterback["Passing Touchdowns"] * scores.iloc[0, 1]
quarterback["Intercepts Thrown Score"] = quarterback["Passing Interceptions"] * scores.iloc[0, 2]
quarterback.head()
```
