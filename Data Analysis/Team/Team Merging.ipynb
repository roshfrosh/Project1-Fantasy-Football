```python
output_data_final = "TeamMerge/merged_team_data_2012_2018.csv"
df1 = pd.merge(touchdown_data_final, rush_data_final, how = 'left', on="Key")
df2 = pd.merge(df1, receiving_data_final, how = 'left', on="Key")
df3 = pd.merge(df2, punts_data_final, how = 'left', on="Key")
df4 = pd.merge(df3, puntreturns_data_final, how = 'left', on="Key")
df5 = pd.merge(df4, penalties_data_final, how = 'left', on="Key")
df6 = pd.merge(df5, passing_data_final, how = 'left', on="Key")
df7 = pd.merge(df6, kickreturns_data_final, how = 'left', on="Key")
df8 = pd.merge(df7, kickoffs_data_final, how = 'left', on="Key")
df9 = pd.merge(df8, interceptions_data_final, how = 'left', on="Key")
df10 = pd.merge(df9, interceptionreturns_data_final, how = 'left', on="Key")
df11 = pd.merge(df10, gamesplayed_data_final, how = 'left', on="Key")
df12 = pd.merge(df11, fumbles_data_final, how = 'left', on="Key")
df13 = pd.merge(df12, firstdowns_data_final, how = 'left', on="Key")
df14 = pd.merge(df13, fieldgoals_data_final, how = 'left', on="Key")
df15 = pd.merge(df14, extrapoints_data_final, how = 'left', on="Key")
df16 = pd.merge(df15, efficiency_data_final, how = 'left', on="Key")
df17 = pd.merge(df16, defense_data_final, how = 'left', on="Key")
merged_data_2012_2018 = pd.merge(df17, conversions_data_final, how = 'left', on="Key")
merged_data_2012_2018.to_csv (output_data_final, index = None, header=True)
merged_data_2012_2018.head()
```
