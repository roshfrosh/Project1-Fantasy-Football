```python
import pandas as pd
import numpy as np
```

```python
data = 'Data10.csv'
df = pd.read_csv(data)
df.columns
```

```python
necessary = df[['Player Name', 'Season', 'Position', 'Rush Red Zone Attempts PG','Passing Redzone Attempts PG', 'Passing Yards FP',
                'Passing Touchdowns FP','Passing Interceptions FP', 'Rush Yards FP', 'Rush Touchdowns FP',
                'Lost Fumbles FP', 'Total FP PG' ]]
necessary.head()
```

```python
quarterback = necessary.loc[necessary["Position"] == "QB"]
quarterback
```

```python
qbno2018 = quarterback[quarterback["Season"] < 2018]
averagesno2018 = qbno2018.groupby("Player Name").mean()
averagesno2018 = averagesno2018[['Rush Red Zone Attempts PG', 'Passing Redzone Attempts PG', "Passing Yards FP",
       'Passing Touchdowns FP', 'Passing Interceptions FP', 'Rush Yards FP',
       'Rush Touchdowns FP', 'Lost Fumbles FP']]
fp2018 = quarterback[quarterback["Season"] == 2018]
fp2018points = fp2018[["Player Name", "Total FP PG"]]
fp2018points
averagesno2018wpoints = pd.merge(averagesno2018, fp2018points, how = 'left', on="Player Name")
averagesno2018wpoints["Predicted Points 2018"] = (averagesno2018wpoints['Rush Red Zone Attempts PG']*.105)\
+(averagesno2018wpoints['Passing Redzone Attempts PG']*1.714)\
+(averagesno2018wpoints['Passing Yards FP']*0.774)\
+(averagesno2018wpoints['Passing Touchdowns FP']*.548)\
+(averagesno2018wpoints['Passing Interceptions FP']*.335)\
+(averagesno2018wpoints['Rush Yards FP']*.61)\
+(averagesno2018wpoints['Rush Touchdowns FP']*.255)\
+(averagesno2018wpoints['Lost Fumbles FP']*-3.033)+2.549
averagesno2018wpoints['Delta'] = averagesno2018wpoints['Total FP PG'] - averagesno2018wpoints['Predicted Points 2018']
averagesno2018wpoints
```

```python
averagesno2018wpoints.columns
top5predicted = averagesno2018wpoints[['Player Name', 'Total FP PG', 'Predicted Points 2018', 'Delta']]
top5predicted = top5predicted.sort_values("Predicted Points 2018", ascending = False)
# top5predicted.rename(columns ={'Total FP FG': 'Actual Fantasy Points 2018'}, inplace=True) 
top5predicted.head(10)
```

```python
top5actual = top5predicted.sort_values("Total FP PG", ascending = False)
top5actual.head(10)
```
