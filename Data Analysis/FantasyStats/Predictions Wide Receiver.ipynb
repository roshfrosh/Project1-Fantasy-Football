```python
import pandas as pd
import numpy as np
```

```python
data = 'Data10.csv'
df = pd.read_csv(data)
df.columns
```

```python
necessary = df[['Player Name', 'Season', 'Position', 'Rush Yards FP',
               'Rush Touchdowns FP','Rush Red Zone Attempts PG', 'Rec Yards FP', 
               'Rec Receptions FP', 'Rec Touchdowns FP', 'Lost Fumbles FP', 'Total FP PG']]
necessary.head()
```

```python
wide = necessary.loc[necessary["Position"] == "WR"]
wide
```

```python
qbno2018 = wide[wide["Season"] < 2018]
averagesno2018 = qbno2018.groupby("Player Name").mean()
averagesno2018
averagesno2018 = averagesno2018[['Rush Yards FP',
               'Rush Touchdowns FP','Rush Red Zone Attempts PG', 'Rec Yards FP', 
               'Rec Receptions FP', 'Rec Touchdowns FP', 'Lost Fumbles FP']]

fp2018 = wide[wide["Season"] == 2018]
fp2018points = fp2018[["Player Name", "Total FP PG"]]
fp2018points
averagesno2018wpoints = pd.merge(averagesno2018, fp2018points, how = 'left', on="Player Name")
averagesno2018wpoints
averagesno2018wpoints["Predicted Points 2018"] = (averagesno2018wpoints['Rush Yards FP']*.054)\
+(averagesno2018wpoints['Rush Touchdowns FP']*-0.248)\
+(averagesno2018wpoints['Rush Red Zone Attempts PG']*10.828)\
+(averagesno2018wpoints['Rec Yards FP']*.574)\
+(averagesno2018wpoints['Rec Receptions FP']*1.401)\
+(averagesno2018wpoints['Rec Touchdowns FP']*.327)\
+(averagesno2018wpoints['Lost Fumbles FP']*-.49) +1.663
averagesno2018wpoints['Delta'] = averagesno2018wpoints['Total FP PG'] - averagesno2018wpoints['Predicted Points 2018']
averagesno2018wpoints
```

```python
averagesno2018wpoints.columns
top5predicted = averagesno2018wpoints[['Player Name', 'Total FP PG', 'Predicted Points 2018', 'Delta']]
top5predicted = top5predicted.sort_values("Predicted Points 2018", ascending = False)
top5predicted.head(10)
```

```python
top5actual = top5predicted.sort_values("Total FP PG", ascending = False)
top5actual.head(10)
```
