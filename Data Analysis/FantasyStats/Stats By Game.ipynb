```python
import pandas as pd
import numpy as np
```

```python
data = 'fantasystatsall.csv'
df = pd.read_csv(data)
df.columns
```

```python
df['Rec Receptions PG'] = df['Rec Receptions']/df['Games Played']   
df['Rec Red Zone Targets PG'] = df['Rec Red Zone Targets']/df['Games Played'] 
df['Rec Touchdowns PG'] = df['Rec Touchdowns']/df['Games Played'] 
df['Rec Yards PG'] = df['Rec Yards']/df['Games Played']  
df['Fumbles PG'] = df['Fumbles']/df['Games Played']  
df['Lost Fumbles PG'] = df['Lost Fumbles']/df['Games Played']  
df['Rush Attempts PG'] = df['Rush Attempts']/df['Games Played']  
df['Rush Avg Yards PG'] = df['Rush Avg Yards']/df['Games Played']  
df['Rush Red Zone Attempts PG'] = df['Rush Red Zone Attempts']/df['Games Played']  
df['Rush Scrambles PG'] = df['Rush Scrambles']/df['Games Played']  
df['Rush Touchdowns PG'] = df['Rush Touchdowns']/df['Games Played']  
df['Rush Yards PG'] = df['Rush Yards']/df['Games Played']  
df['Passing Completion Pct PG'] = df['Passing Completion Pct']/df['Games Played']  
df['Passing Hurries PG'] = df['Passing Hurries']/df['Games Played']  
df['Passing Interceptions PG'] = df['Passing Interceptions']/df['Games Played']  
df['Passing Redzone Attempts PG'] = df['Passing Redzone Attempts']/df['Games Played']  
df['Passing Touchdowns PG'] = df['Passing Touchdowns']/df['Games Played']  
df['Passing Yards PG'] = df['Passing Yards']/df['Games Played']  
df['Field Goal Attempts PG'] = df['Field Goal Attempts']/df['Games Played']  
df['Field Goal Longest PG'] = df['Field Goal Longest']/df['Games Played']  
df['Field Goal Made PG'] = df['Field Goal Made']/df['Games Played']  
df['Field Goal Yards PG'] = df['Field Goal Yards']/df['Games Played']  
df['Extra Pts Attempts PG'] = df['Extra Pts Attempts']/df['Games Played']  
df['Extra Pts Made PG'] = df['Extra Pts Made']/df['Games Played']  
df.head()
```
