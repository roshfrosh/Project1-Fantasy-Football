```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import sys

key = "ja987zj97xrfhjmdfu563wv4"
```

```python
# File to Load 
team_data_to_load = "Resources/Input/Team_List.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
```

```python
#Receiving Data
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
team_list = team_df["Team ID"].tolist()


#lists
team_id_df = []
season_df = []
key_df = []
sos_sos_df= []
sos_wins_df = []
sos_totals_df = []
sov_sov_df= []
sov_wins_df=[]
sov_totals_df=[]

#params to process player descriptive data
for a in range(len(year_list)):
    try:
        yr = year_list[a]
        url = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/"  + yr + "/standings.json?api_key=" + key
        time.sleep(2)
        JSONContent = requests.get(url).json()
        conf_ct = len(JSONContent["conferences"])
        print(conf_ct)

        for i in range (0, conf_ct):
            div_ct = len(JSONContent["conferences"][i]) 
            for j in range (0, div_ct):
                team_ct = len(JSONContent["conferences"][i]["divisions"][j]["teams"])
                for k in range (0, team_ct):
                    try:
                        team_id = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["id"]
                        sos_sos = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["strength_of_schedule"]["sos"]
                        sos_wins = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["strength_of_schedule"]["wins"]
                        sos_totals = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["strength_of_schedule"]["total"]
                        sov_sov = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["strength_of_victory"]["sov"]
                        sov_wins = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["strength_of_victory"]["wins"]
                        sov_totals = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["strength_of_victory"]["total"]
                        #team_id = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["id"]
                        season_df.append(yr)
                        team_id_df.append(team_id)
                        key_df.append(team_id+yr)
                        sos_sos_df.append(sos_sos)
                        sos_wins_df.append(sos_wins)
                        sos_totals_df.append(sos_totals)
                        sov_sov_df.append(sov_sov)
                        sov_wins_df.append(sov_wins)
                        sov_totals_df.append(sov_totals)
                    except:
                        print("Team Fail: "+ team_id)
                        print(sys.exc_info())  
                        continue 
    except:
        print("Year Fail: "+ yr)
        print(sys.exc_info())  
        continue 

    

data = {"Season": season_df, 
        "Team ID": team_id_df, 
        "TKey": key_df, 
        "SOS": sos_sos_df,
        "SOS Wins": sos_wins_df,
        "SOS Totals": sos_totals_df,
        "SOV": sov_sov_df,
        "SOV Wins":sov_wins_df,
        "SOV Totals": sov_totals_df,
       }
                       
                    
df = pd.DataFrame(data) 
df.head()
```
