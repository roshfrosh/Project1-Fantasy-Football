```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import sys

key = "va625m7eag2pgv8ydupypvdn"
```

```python
# File to Load 
team_data_to_load = "Resources/Input/Team_List.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
```

```python
#Receiving Data
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
team_list = team_df["Team ID"].tolist()


#lists
team_id_df = []
season_df = []
key_df = []
attempt_3d_df=[]
succ_3d_df=[]
pct_3d_df=[]
attempt_4d_df=[]
succ_4d_df=[]
pct_4d_df=[]
attempt_gtg_df=[]
succ_gtg_df=[]
pct_gtg_df=[]
attempt_rz_df=[]
succ_rz_df=[]
pct_rz_df=[]




#params to process player descriptive data
for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(2)
        JSONContent = requests.get(url).json()
        try: 
            attempt_3d=JSONContent["record"]["efficiency"]["thirddown"]["attempts"]
            succ_3d=JSONContent["record"]["efficiency"]["thirddown"]["successes"] 
            pct_3d=JSONContent["record"]["efficiency"]["thirddown"]["pct"] 
            attempt_4d=JSONContent["record"]["efficiency"]["fourthdown"]["attempts"]  
            succ_4d=JSONContent["record"]["efficiency"]["fourthdown"]["successes"] 
            pct_4d=JSONContent["record"]["efficiency"]["fourthdown"]["pct"] 
            attempt_gtg=JSONContent["record"]["efficiency"]["goaltogo"]["attempts"] 
            succ_gtg=JSONContent["record"]["efficiency"]["goaltogo"]["successes"] 
            pct_gtg=JSONContent["record"]["efficiency"]["goaltogo"]["pct"] 
            attempt_rz=JSONContent["record"]["efficiency"]["redzone"]["attempts"] 
            succ_rz=JSONContent["record"]["efficiency"]["redzone"]["successes"] 
            pct_rz=JSONContent["record"]["efficiency"]["redzone"]["pct"] 
            season_df.append(yr)
            team_id_df.append(team_list[i])
            key_df.append(team_list[i]+yr)
            attempt_3d_df.append(attempt_3d)
            succ_3d_df.append(succ_3d)
            pct_3d_df.append(pct_3d)
            attempt_4d_df.append(attempt_4d)
            succ_4d_df.append(succ_4d)
            pct_4d_df.append(pct_4d)
            attempt_gtg_df.append(attempt_gtg)
            succ_gtg_df.append(succ_gtg)
            pct_gtg_df.append(pct_gtg)
            attempt_rz_df.append(attempt_rz)
            succ_rz_df.append(succ_rz)
            pct_rz_df.append(pct_rz)
            print("Team: "+team_list[i])
        except:
            print("Team Fail: "+team_list[i])
            print(sys.exc_info())  
            continue 

data = {"Season": season_df, 
        "Team ID": team_id_df, 
        "TKey": key_df, 
        "3D Attempt":attempt_3d_df,
        "3D Successes":succ_3d_df,
        "3D Pct":pct_3d_df,
        "4D Attempt":attempt_4d_df,
        "4D Successes":succ_4d_df,
        "4D Pct":pct_4d_df,
        "Goal to Go Attempts":attempt_gtg_df,
        "Goal to Go Successes":succ_gtg_df,
        "Goal to Go Pct":pct_gtg_df,
        "RedZone Attempts":attempt_rz_df,
        "RedZone Successes":succ_rz_df,
        "RedZone Pct":pct_rz_df,
       }

df = pd.DataFrame(data) 
df.head()
```
