```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import sys

key = "kvnrxwq55ysrkb6295gtus2r"
```

```python
# File to Load 
team_data_to_load = "Resources/Input/Team_List.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
```

```python
#Receiving Data
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
team_list = team_df["Team ID"].tolist()


#lists
team_id_df = []
season_df = []
key_df = []
pass_air_yds_df =[]
pass_attempts_df =[]
pass_avg_pocket_time_df =[]
pass_avg_poor_throws_df = []
pass_avg_yds_df =[]
pass_blitzes_df =[]
pass_completion_pct_df =[] 
pass_completions_df =[]
pass_defended_passes_df =[]
pass_dropped_passes_df =[]
pass_gross_yards_df =[]
pass_hurries_df =[]
pass_interceptions_df =[]
pass_knockdowns_df =[]
pass_longest_df =[]
pass_longest_td_df =[]
pass_net_yds_df =[]
pass_pocket_time_df = []
pass_rating_df = []
pass_redzone_attempts_df =[]
pass_sack_yds_df =[]
pass_sacks_df =[]
pass_spikes_df =[]
pass_throw_away_df =[]
pass_touchdowns_df =[]
pass_yards_df =[]


#params to process player descriptive data
for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(2)
        JSONContent = requests.get(url).json()
        try: 
            pass_air_yds =JSONContent["record"]["passing"]["air_yards"]
            pass_attempts =JSONContent["record"]["passing"]["attempts"] 
            try:
                pass_avg_pocket_time =JSONContent["record"]["passing"]["avg_pocket_time"]
            except: 
                pass_avg_pocket_time = 0
            try:
                pass_avg_poor_throws=JSONContent["record"]["passing"]["poor_throws"] 
            except: 
                pass_avg_pocket_time = 0
            pass_avg_yds =JSONContent["record"]["passing"]["avg_yards"] 
            pass_blitzes =JSONContent["record"]["passing"]["blitzes"]  
            pass_completion_pct =JSONContent["record"]["passing"]["cmp_pct"] 
            pass_completions =JSONContent["record"]["passing"]["completions"]
            pass_defended_passes =JSONContent["record"]["passing"]["defended_passes"] 
            pass_dropped_passes =JSONContent["record"]["passing"]["dropped_passes"] 
            pass_gross_yards =JSONContent["record"]["passing"]["gross_yards"] 
            pass_hurries =JSONContent["record"]["passing"]["hurries"] 
            pass_interceptions =JSONContent["record"]["passing"]["interceptions"] 
            pass_knockdowns =JSONContent["record"]["passing"]["knockdowns"] 
            pass_longest =JSONContent["record"]["passing"]["longest"] 
            pass_longest_td =JSONContent["record"]["passing"]["longest_touchdown"]
            try:
                pass_net_yds =JSONContent["record"]["passing"]["net_yards"]
            except:
                pass_net_yds =0
            pass_pocket_time = JSONContent["record"]["passing"]["pocket_time"] 
            pass_rating = JSONContent["record"]["passing"]["rating"] 
            pass_redzone_attempts =JSONContent["record"]["passing"]["redzone_attempts"] 
            pass_sack_yds =JSONContent["record"]["passing"]["sack_yards"] 
            pass_sacks =JSONContent["record"]["passing"]["sacks"] 
            pass_spikes =JSONContent["record"]["passing"]["spikes"] 
            pass_throw_aways =JSONContent["record"]["passing"]["throw_aways"] 
            pass_touchdowns =JSONContent["record"]["passing"]["touchdowns"] 
            pass_yards =JSONContent["record"]["passing"]["yards"]
            season_df.append(yr)
            team_id_df.append(team_list[i])
            key_df.append(team_list[i]+yr)
            pass_air_yds_df.append(pass_air_yds)
            pass_attempts_df.append(pass_attempts)
            pass_avg_pocket_time_df.append(pass_avg_pocket_time)
            pass_avg_poor_throws_df.append(pass_avg_poor_throws)
            pass_avg_yds_df.append(pass_avg_yds)
            pass_blitzes_df.append(pass_blitzes)
            pass_completion_pct_df.append(pass_completion_pct)
            pass_completions_df.append(pass_completions)
            pass_defended_passes_df.append(pass_defended_passes)
            pass_dropped_passes_df.append(pass_dropped_passes)
            pass_gross_yards_df.append(pass_gross_yards)
            pass_hurries_df.append(pass_hurries)
            pass_interceptions_df.append(pass_interceptions)
            pass_knockdowns_df.append(pass_knockdowns)
            pass_longest_df.append(pass_longest)
            pass_longest_td_df.append(pass_longest_td)
            pass_net_yds_df.append(pass_net_yds)
            pass_pocket_time_df.append(pass_pocket_time)
            pass_rating_df.append(pass_rating)
            pass_redzone_attempts_df.append(pass_redzone_attempts)
            pass_sack_yds_df.append(pass_sack_yds)
            pass_sacks_df.append(pass_sacks)
            pass_spikes_df.append(pass_spikes)
            pass_throw_away_df.append(pass_throw_aways)
            pass_touchdowns_df.append(pass_touchdowns)
            pass_yards_df.append(pass_yards)
            print("Team: "+team_list[i])
        except:
            print("Team Fail: "+team_list[i])
            print(sys.exc_info())  
            continue 

data = {"Season": season_df, 
        "Team ID": team_id_df, 
        "TKey": key_df, 
        "Pass Air Yds":pass_air_yds_df,
        "Pass Attempts":pass_attempts_df,
        "Pass Avg Pocket Time":pass_avg_pocket_time_df,
        "Pass Poor Throws": pass_avg_poor_throws_df,
        "Passing Avg Yards": pass_avg_yds_df,
        "Passing Blitzes": pass_blitzes_df,
        "Passing Completion Pct": pass_completion_pct_df,
        "Passing Completions": pass_completions_df,
        "Passing Defended Passes": pass_defended_passes_df,
        "Passing Dropped Passes": pass_dropped_passes_df,
        "Passing Gross Yards": pass_gross_yards_df,
        "Passing Hurries": pass_hurries_df,
        "Passing Interceptions": pass_interceptions_df,
        "Passing Knockdowns": pass_knockdowns_df,
        "Passing Longest": pass_longest_df,
        "Passing Longest TD": pass_longest_td_df,
        "Passing Net Yds": pass_net_yds_df,
        "Passing Pocket Time": pass_pocket_time_df,
        "Passing Rating": pass_rating_df,
        "Passing Redzone Attempts": pass_redzone_attempts_df,
        "Passing Sack Yds": pass_sack_yds_df,
        "Passing Sacks": pass_sacks_df,
        "Passing Spikes": pass_spikes_df,
        "Passing Throw Away": pass_throw_away_df,
        "Passing Touchdowns": pass_touchdowns_df,
        "Passing Yards": pass_yards_df
       }

df = pd.DataFrame(data) 
df.head()
```
