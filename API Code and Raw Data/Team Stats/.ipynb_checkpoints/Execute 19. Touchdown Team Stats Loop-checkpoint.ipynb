```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import sys

key = "etkq9nhbzgb46xeqy4qasax6"
```

```python
# File to Load 
team_data_to_load = "Resources/Input/Team_List.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
```

```python
#Receiving Data
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
team_list = team_df["Team ID"].tolist()


#lists
team_id_df = []
season_df = []
key_df = []
td_fum_ret_df = []
td_int_ret_df = []
td_kick_ret_df = []
td_other_df = []
td_pass_df =[]
td_punt_ret_df =[]
td_rush_df =[]
td_total_df =[] 
td_total_ret_df =[]

#params to process player descriptive data
for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(2)
        JSONContent = requests.get(url).json()
        try: 
            td_fum_ret = JSONContent["record"]["touchdowns"]["fumble_return"]  
            td_int_ret = JSONContent["record"]["touchdowns"]["int_return"] 
            td_kick_ret = JSONContent["record"]["touchdowns"]["kick_return"] 
            td_other = JSONContent["record"]["touchdowns"]["other"] 
            td_pass =JSONContent["record"]["touchdowns"]["pass"] 
            td_punt_ret =JSONContent["record"]["touchdowns"]["punt_return"] 
            td_rush =JSONContent["record"]["touchdowns"]["rush"]
            td_total =JSONContent["record"]["touchdowns"]["total"] 
            td_total_ret =JSONContent["record"]["touchdowns"]["total_return"] 
            td_fum_ret_df.append(td_fum_ret)
            td_int_ret_df.append(td_int_ret)
            td_kick_ret_df.append(td_kick_ret)
            td_other_df.append(td_other)
            td_pass_df.append(td_pass)
            td_punt_ret_df.append(td_punt_ret)
            td_rush_df.append(td_rush)
            td_total_df.append(td_total) 
            td_total_ret_df.append(td_total_ret)
            season_df.append(yr)
            team_id_df.append(team_list[i])
            key_df.append(team_list[i]+yr)
            print("Team: "+team_list[i])
        except:
            print("Team Fail: "+team_list[i])
            print(sys.exc_info())  
            continue 

data = {"Season": season_df, 
        "Team ID": team_id_df, 
        "TKey": key_df, 
        "TD Fumble Returns":td_fum_ret_df,
        "TD Interception Return":td_int_ret_df,
        "TD Kick Return":td_kick_ret_df,
        "TD Other":td_other_df,
        "TD Pass":td_pass_df,
        "TD Punt Return":td_punt_ret_df,
        "TD Rush":td_rush_df,
        "TD Total":td_total_df,
        "TD Total Return":td_total_ret_df

       }

df = pd.DataFrame(data) 
df.head()
```

```python
df.to_csv(r'Resources/Output/Team_Stats_2012_2018_Touchdowns.csv')
```
