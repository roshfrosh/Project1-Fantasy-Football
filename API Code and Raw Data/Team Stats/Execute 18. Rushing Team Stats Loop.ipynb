```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import sys

key = "gfwcjzfwvhhc8t7ytnfpuzz3"
```

```python
# File to Load 
team_data_to_load = "Resources/Input/Team_List.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
```

```python
#Receiving Data
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
team_list = team_df["Team ID"].tolist()


#lists
team_id_df = []
season_df = []
key_df = []
rush_attempts_df= []
rush_avg_yards_df = []
rush_broken_tackles_df= [] 
rush_kneel_downs_df= [] 
rush_longest_df= []
rush_longest_td_df=[]
rush_rz_attempts_df=[]
rush_scrambles_df=[]
rush_tlost_df=[]
rush_tlost_yds_df=[]
rush_tds_df=[]
rush_yds_df=[]
rush_yds_after_cont_df=[]

#params to process player descriptive data
for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(2)
        JSONContent = requests.get(url).json()
        try: 
            rush_attempts= JSONContent["record"]["rushing"]["attempts"]  
            rush_avg_yards = JSONContent["record"]["rushing"]["avg_yards"]  
            rush_broken_tackles= JSONContent["record"]["rushing"]["broken_tackles"]  
            rush_kneel_downs= JSONContent["record"]["rushing"]["kneel_downs"]  
            rush_longest= JSONContent["record"]["rushing"]["longest"]  
            rush_longest_td=JSONContent["record"]["rushing"]["longest_touchdown"]  
            rush_rz_attempts=JSONContent["record"]["rushing"]["redzone_attempts"]  
            rush_scrambles=JSONContent["record"]["rushing"]["scrambles"]  
            rush_tlost=JSONContent["record"]["rushing"]["tlost"]  
            rush_tlost_yds=JSONContent["record"]["rushing"]["tlost_yards"]  
            rush_tds=JSONContent["record"]["rushing"]["touchdowns"]  
            rush_yds=JSONContent["record"]["rushing"]["yards"]  
            rush_yds_after_cont=JSONContent["record"]["rushing"]["yards_after_contact"]
            rush_attempts_df.append(rush_attempts)
            rush_avg_yards_df.append(rush_avg_yards)
            rush_broken_tackles_df.append(rush_broken_tackles)
            rush_kneel_downs_df.append(rush_kneel_downs)
            rush_longest_df.append(rush_longest)
            rush_longest_td_df.append(rush_longest_td)
            rush_rz_attempts_df.append(rush_rz_attempts)
            rush_scrambles_df.append(rush_scrambles)
            rush_tlost_df.append(rush_tlost)
            rush_tlost_yds_df.append(rush_tlost_yds)
            rush_tds_df.append(rush_tds)
            rush_yds_df.append(rush_yds)
            rush_yds_after_cont_df.append(rush_yds_after_cont)
            season_df.append(yr)
            team_id_df.append(team_list[i])
            key_df.append(team_list[i]+yr)
            print("Team: "+team_list[i])
        except:
            print("Team Fail: "+team_list[i])
            print(sys.exc_info())  
            continue 

data = {"Season": season_df, 
        "Team ID": team_id_df, 
        "TKey": key_df, 
        "Rush Attempts": rush_attempts_df,
        "Rush Avg Yards":rush_avg_yards_df,
        "Rush Broken Tackles":rush_broken_tackles_df,
        "Rush Kneel Downs":rush_kneel_downs_df,
        "Rush Longest":rush_longest_df,
        "Rush Longest TD":rush_longest_td_df,
        "Rush Red Zone Attempts":rush_rz_attempts_df,
        "Rush Scrambles":rush_scrambles_df,
        "Rush Tackles for Loss":rush_tlost_df,
        "Rush Tackles for Loss Yards":rush_tlost_yds_df,
        "Rush Touchdowns":rush_tds_df,
        "Rush Yards":rush_yds_df,
        "Rush Yards After Contact":rush_yds_after_cont_df
       }

df = pd.DataFrame(data) 
df.head()
```
