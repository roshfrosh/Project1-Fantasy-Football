```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from datetime import datetime
from dateutil.parser import parse
```

```python
# File to Load 
team_data_to_load = "Team_List.csv"
players_2018_data_to_load = "Offense_Player_List_2018.csv"
player_bday_to_load = "Player_List_Bday.csv"
player_2012to2018_to_load = "Player_List_2012_to_2018.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
players_2018_data = pd.read_csv(players_2018_data_to_load)
player_bday_data =  pd.read_csv(player_bday_to_load)
big_player_list =  pd.read_csv(player_2012to2018_to_load)
```

```python
player_bday_data .head()
```

```python
big_player_list.head()

result = pd.merge(big_player_list,
                 player_bday_data[["Birth Date", "Player ID"]],
                 on="Player ID")
result.head()

bday_list = result["Birth Date"].tolist()
season_list = result["Season"].tolist()

age_list = []

for i in range(len(bday_list)):
    if bday_list[i] == "none": 
        age_list.append("unknown")
    else: 
        age= season_list[i] - datetime.strptime(bday_list[i], '%Y-%m-%d').year
        age_list.append(age)

result["Season Age"] = age_list
    
result.head()
    
```

```python
result.to_csv(r'Player_List_2012_to_2018_WITH_Player_Age.csv')
```
