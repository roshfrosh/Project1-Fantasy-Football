```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import sys

key = "jxt3v9r9gdbvxhxcvarr7uc6"
```

```python
# File to Load 
team_data_to_load = "Team_List.csv"
players_2018_data_to_load = "Offense_Player_List_2018.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
players_2018_data = pd.read_csv(players_2018_data_to_load)
```

```python
#Conversions
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
offensive_pos = ["TE", "WR", "RB", "QB", "FB", "K"]
master_player_list = players_2018_data["Player ID"].tolist()
team_list = team_df["Team ID"].tolist()
teamname_list = team_df["Name"].tolist()


pid_df = []
season_df = []
fg_attempts_df =[]
fg_avg_yards_df =[] 
fg_blocked_df =[] 
fg_longest_df =[] 
fg_made_df =[]
fg_yards_df =[] 



for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url2 = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(1)
        JSONContent2 = requests.get(url2).json()
        player_ct = len(JSONContent2["players"])
        for j in range (0, player_ct):
            pid = JSONContent2["players"][j]["id"]
            if pid in master_player_list:
                try:
                    fg_attempts =JSONContent2["players"][j]["field_goals"]["attempts"]  
                    fg_avg_yards =JSONContent2["players"][j]["field_goals"]["avg_yards"]  
                    fg_blocked =JSONContent2["players"][j]["field_goals"]["blocked"]  
                    fg_longest =JSONContent2["players"][j]["field_goals"]["longest"]  
                    fg_made =JSONContent2["players"][j]["field_goals"]["made"]  
                    fg_yards =JSONContent2["players"][j]["field_goals"]["yards"]  
                    season_df.append(yr)
                    pid_df.append(pid)
                    fg_attempts_df.append(fg_attempts)
                    fg_avg_yards_df.append(fg_avg_yards)
                    fg_blocked_df.append(fg_blocked) 
                    fg_longest_df.append(fg_longest)
                    fg_made_df.append(fg_made)
                    fg_yards_df.append(fg_yards)  


                except: 
                    print(sys.exc_info())  
                    continue 


data5 = {"Player Season": season_df,
        "Player ID": pid_df,
        "Field Goal Attempts":fg_attempts_df,
        "Field Goal Avg Yards":fg_avg_yards_df,
        "Field Goal Blocked":fg_blocked_df,
        "Field Goal Longest":fg_longest_df,
        "Field Goal Made":fg_made_df,
        "Field Goal Yards":fg_yards_df  

      }

player_fg_df = pd.DataFrame(data5) 
player_fg_df.head()

```
