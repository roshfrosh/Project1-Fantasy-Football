```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

key = "6k7mk9522fq7vrms75sac23y"
```

```python
# File to Load 
team_data_to_load = "Team_List.csv"
players_2018_data_to_load = "Offense_Player_List_2018.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
players_2018_data = pd.read_csv(players_2018_data_to_load)
```

```python
#Build player dataframe for players 2012 - 2018

#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
offensive_pos = ["TE", "WR", "RB", "QB", "FB", "K"]
master_player_list = players_2018_data["Player ID"].tolist()
team_list = team_df["Team ID"].tolist()
teamname_list = team_df["Name"].tolist()


#params to process player descriptive data
pID_df = []
pname_df = []
pjersey_df = []
preference_df = []
pposition_df = []
psrid_df = []
gplayed_df = []
gstarted_df = []
season_df = []
pteamid_df = []
pteamname_df = []

for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        url2 = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        print (str(i) + "||" + yr)
        time.sleep(2)
        JSONContent2 = requests.get(url2).json()
        for j in range (len(JSONContent2["players"])):
            try: 
                pposition = JSONContent2["players"][j]["position"]
                pid = JSONContent2["players"][j]["id"]
                if pposition in offensive_pos:
                    if pid in master_player_list:  
                        pteamid = team_list[i]
                        pteamid_df.append(pteamid)
                        pteamname = teamname_list[i]
                        pteamname_df.append(pteamname)
                        pid = JSONContent2["players"][j]["id"]
                        pID_df.append(pid)
                        pname = JSONContent2["players"][j]["name"]
                        pname_df.append(pname)
                        pjersey = JSONContent2["players"][j]["jersey"]
                        pjersey_df.append(pjersey)
                        preference = JSONContent2["players"][j]["reference"]
                        preference_df.append(preference)
                        pposition = JSONContent2["players"][j]["position"]
                        pposition_df.append(pposition)
                        psrid = JSONContent2["players"][j]["sr_id"]
                        psrid_df.append(psrid)
                        gplayed = JSONContent2["players"][j]["games_played"]
                        gplayed_df.append(gplayed)
                        gstarted = JSONContent2["players"][j]["games_started"]
                        gstarted_df.append(gstarted)
                        season_df.append(yr)
            except: 
                continue


data = {"Player Team ID": pteamid_df , "Player Team": pteamname_df, "Player ID": pID_df,"Player Name": pname_df, 
        "Jersey": pjersey_df, "Reference": preference_df, "Position": pposition_df, "Sr ID": psrid_df, 
        "Games Played": gplayed_df, "Games Started": gstarted_df, "Season": season_df}
player_df = pd.DataFrame(data) 
print("done")
```
