```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

key = "b7q2wzzxzf87t28yqwyxtjvy"
```

```python
# File to Load 
team_data_to_load = "Team_List.csv"
players_2018_data_to_load = "Offense_Player_List_2018.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
players_2018_data = pd.read_csv(players_2018_data_to_load)
```

```python
#kick returns stats
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
offensive_pos = ["TE", "WR", "RB", "QB", "FB", "K"]
master_player_list = players_2018_data["Player ID"].tolist()
team_list = team_df["Team ID"].tolist()
teamname_list = team_df["Name"].tolist()


kryards_df = [] 
kravgyards_df = []
krfaircatches_df = []
krlongest_df = []
krlongesttd_df = []
krretruns_df = []
krtds_df = []
pid_df = []
season_df = []

for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url5 = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(2)
        JSONContent5 = requests.get(url5).json()
        for j in range (len(JSONContent5["players"])):
            pid = JSONContent5["players"][j]["id"]
            if pid in master_player_list:
                try:
                    kryards = JSONContent5["players"][j]["kick_returns"]["yards"] 
                    kravgyards = JSONContent5["players"][j]["kick_returns"]["avg_yards"] 
                    krfaircatches = JSONContent5["players"][j]["kick_returns"]["faircatches"] 
                    krlongest = JSONContent5["players"][j]["kick_returns"]["longest"] 
                    krlongesttd = JSONContent5["players"][j]["kick_returns"]["longest_touchdown"]
                    krretruns = JSONContent5["players"][j]["kick_returns"]["returns"]
                    krtds = JSONContent5["players"][j]["kick_returns"]["touchdowns"]
                    kryards_df.append(kryards)
                    kravgyards_df.append(kravgyards)
                    krfaircatches_df.append(krfaircatches)
                    krlongest_df.append(krlongest)
                    krlongesttd_df.append(krlongesttd)
                    krretruns_df.append(krretruns)
                    krtds_df.append(krtds)
                    season_df.append(yr)
                    pid_df.append(pid)
                    print(pid)
                except:
                   continue


data4 = {"Player ID": pid_df,
        "Season": season_df,
        "Kick Return Yards": kryards_df,
        "Kick Return Avg Yards":kravgyards_df,
        "Kick Return Fair Catches":krfaircatches_df,
        "Kick Return Longest":krlongest_df,
        "Kick Return Longest":krlongesttd_df,
        "Kick Return Returns":krretruns_df,
        "Kick Return Touchdowns":krtds_df       
      }

player_kr_df = pd.DataFrame(data4) 
player_kr_df.head()

```
