```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import sys

key = "hqv6k9jfj8qtp3hxj4c7t2pn"
```

```python
# File to Load 
team_data_to_load = "Team_List.csv"
players_2018_data_to_load = "Offense_Player_List_2018.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
players_2018_data = pd.read_csv(players_2018_data_to_load)
```

```python
#Conversions
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
offensive_pos = ["TE", "WR", "RB", "QB", "FB", "K"]
master_player_list = players_2018_data["Player ID"].tolist()
team_list = team_df["Team ID"].tolist()
teamname_list = team_df["Name"].tolist()


pid_df = []
season_df = []
ko_ez_df = []
ko_inside_20_df = []
ko_kickoffs_df = [] 
ko_onside_attempts_df = []
ko_onside_successes_df = []
ko_oob_df = []
ko_return_yds_df = []
ko_squib_kicks_df = []
ko_touchbacks_df = []
ko_yds_df = []




for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url2 = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(1)
        JSONContent2 = requests.get(url2).json()
        player_ct = len(JSONContent2["players"])
        for j in range (0, player_ct):
            pid = JSONContent2["players"][j]["id"]
            if pid in master_player_list:
                try:
                    ko_ez = JSONContent2["players"][j]["kickoffs"]["endzone"] 
                    ko_inside_20 = JSONContent2["players"][j]["kickoffs"]["inside_20"] 
                    ko_kickoffs = JSONContent2["players"][j]["kickoffs"]["kickoffs"] 
                    ko_onside_attempts = JSONContent2["players"][j]["kickoffs"]["onside_attempts"] 
                    ko_onside_successes = JSONContent2["players"][j]["kickoffs"]["onside_successes"] 
                    ko_oob = JSONContent2["players"][j]["kickoffs"]["out_of_bounds"] 
                    ko_return_yds = JSONContent2["players"][j]["kickoffs"]["return_yards"] 
                    ko_squib_kicks = JSONContent2["players"][j]["kickoffs"]["squib_kicks"] 
                    ko_touchbacks = JSONContent2["players"][j]["kickoffs"]["touchbacks"] 
                    ko_yds = JSONContent2["players"][j]["kickoffs"]["yards"] 
                    season_df.append(yr)
                    pid_df.append(pid)
                    ko_ez_df.append(ko_ez)
                    ko_inside_20_df.append(ko_inside_20)
                    ko_kickoffs_df.append(ko_kickoffs)
                    ko_onside_attempts_df.append(ko_onside_attempts)
                    ko_onside_successes_df.append(ko_onside_successes)
                    ko_oob_df.append(ko_oob)
                    ko_return_yds_df.append(ko_return_yds)
                    ko_squib_kicks_df.append(ko_squib_kicks)
                    ko_touchbacks_df.append(ko_touchbacks)
                    ko_yds_df.append(ko_yds)
                except: 
                    print(sys.exc_info())  
                    continue 


data5 = {"Player Season": season_df,
        "Player ID": pid_df,
        "Kick Off End Zone":ko_ez_df,
        "Kick Off Inside 20":ko_inside_20_df,
        "Kick Off Kick Offs":ko_kickoffs_df,
        "Kick Off Onside Attempts":ko_onside_attempts_df,
        "Kick Off Onside Successes":ko_onside_successes_df,
        "Kick Off OOB":ko_oob_df,
        "Kick Off Ret Yds":ko_return_yds_df,
        "Kick Off Squib Kicks":ko_squib_kicks_df,
        "Kick Off Touchbacks":ko_touchbacks_df,
        "Kick Off Yds":ko_yds_df

      }

player_ko_df = pd.DataFrame(data5) 
player_ko_df.head()

```
