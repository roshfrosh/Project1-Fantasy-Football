```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

key = "3tqkcnf7hs5wvwtryxyy2jru"
```

```python
# File to Load 
team_data_to_load = "Team_List.csv"
players_2018_data_to_load = "Offense_Player_List_2018.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
players_2018_data = pd.read_csv(players_2018_data_to_load)
```

```python
#Receiving Data
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
offensive_pos = ["TE", "WR", "RB", "QB", "FB", "K"]
master_player_list = players_2018_data["Player ID"].tolist()
team_list = team_df["Team ID"].tolist()
teamname_list = team_df["Name"].tolist()

#lists
rec_air_yds_df = []
rec_avg_yds_df = []
rec_broken_tackles_df = []
rec_catch_passes_df =[] 
rec_dropped_passes_df =[]
rec_longest_df =[]
rec_longest_td_df =[]
rec_receptions_df =[]
rec_rz_targets_df =[]
rec_targets_df =[]
rec_tds_df =[]
rec_yds_df =[]
rec_yds_ac_df =[]
rec_yds_actc_df =[]
pid_df = []
season_df = []

#params to process player descriptive data
for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url3 = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(2)
        JSONContent3 = requests.get(url3).json()
        for j in range (len(JSONContent3["players"])):
            pid = JSONContent3["players"][j]["id"]
            if pid in master_player_list: 
                try: 
                    rec_air_yds = JSONContent3["players"][j]["receiving"]["air_yards"]
                    rec_air_yds_df.append(rec_air_yds)
                    rec_avg_yds =JSONContent3["players"][j]["receiving"]["avg_yards"]
                    rec_avg_yds_df.append(rec_avg_yds)
                    rec_broken_tackles =JSONContent3["players"][j]["receiving"]["broken_tackles"]
                    rec_broken_tackles_df.append(rec_broken_tackles)
                    rec_catch_passes =JSONContent3["players"][j]["receiving"]["catchable_passes"] 
                    rec_catch_passes_df.append(rec_catch_passes)
                    rec_dropped_passes =JSONContent3["players"][j]["receiving"]["dropped_passes"] 
                    rec_dropped_passes_df.append(rec_dropped_passes)
                    rec_longest =JSONContent3["players"][j]["receiving"]["longest"] 
                    rec_longest_df.append(rec_longest)
                    rec_longest_td =JSONContent3["players"][j]["receiving"]["longest_touchdown"] 
                    rec_longest_td_df.append(rec_longest_td)
                    rec_receptions =JSONContent3["players"][j]["receiving"]["receptions"]
                    rec_receptions_df.append(rec_receptions)
                    rec_rz_targets =JSONContent3["players"][j]["receiving"]["redzone_targets"] 
                    rec_rz_targets_df.append(rec_rz_targets)
                    rec_targets =JSONContent3["players"][j]["receiving"]["targets"]
                    rec_targets_df.append(rec_targets)
                    rec_tds =JSONContent3["players"][j]["receiving"]["touchdowns"]
                    rec_tds_df.append(rec_tds)
                    rec_yds =JSONContent3["players"][j]["receiving"]["yards"] 
                    rec_yds_df.append(rec_yds)
                    rec_yds_ac =JSONContent3["players"][j]["receiving"]["yards_after_catch"] 
                    rec_yds_ac_df.append(rec_yds_ac)
                    rec_yds_actc =JSONContent3["players"][j]["receiving"]["yards_after_contact"]
                    rec_yds_actc_df.append(rec_yds_actc)
                    season_df.append(yr)
                    pid_df.append(pid)
                except:
                    rec_air_yds = 0
                    rec_avg_yds =0
                    rec_broken_tackles =0 
                    rec_catch_passes =0 
                    rec_dropped_passes =0
                    rec_longest =0 
                    rec_longest_td =0
                    rec_receptions =0
                    rec_rz_targets =0
                    rec_targets =0
                    rec_tds =0
                    rec_yds =0
                    rec_yds_ac =0
                    rec_yds_actc =0
                    rec_air_yds_df.append(rec_air_yds)
                    rec_avg_yds_df.append(rec_avg_yds)
                    rec_broken_tackles_df.append(rec_broken_tackles)
                    rec_catch_passes_df.append(rec_catch_passes)
                    rec_dropped_passes_df.append(rec_dropped_passes)
                    rec_longest_df.append(rec_longest)
                    rec_longest_td_df.append(rec_longest_td)
                    rec_receptions_df.append(rec_receptions)
                    rec_rz_targets_df.append(rec_rz_targets)
                    rec_targets_df.append(rec_targets)
                    rec_tds_df.append(rec_tds)
                    rec_yds_df.append(rec_yds)
                    rec_yds_ac_df.append(rec_yds_ac)
                    rec_yds_actc_df.append(rec_yds_actc)
                    season_df.append(yr)
                    pid_df.append(pid)

data2 = {"Player ID": pid_df , 
        "Season": season_df, "Rec Air Yards": rec_air_yds_df,
        "Rec Avg Yards": rec_avg_yds_df, "Rec Broken Tackles": rec_broken_tackles_df, 
        "Rec Catchable Passes": rec_catch_passes_df, "Rec Dropped Passes": rec_dropped_passes_df, 
        "Rec Longest": rec_longest_df, "Rec Longest TD": rec_longest_td_df,
        "Rec Receptions": rec_receptions_df, "Rec Red Zone Targets": rec_rz_targets_df, 
        "Rec Targets": rec_targets_df, "Rec Touchdowns": rec_tds_df, 
        "Rec Yards": rec_yds_df, "Rec Yards After Catch": rec_yds_ac_df,
        "Rec Yards After Contact": rec_yds_actc_df}

player_rec_df = pd.DataFrame(data2) 
player_rec_df.head()
```
