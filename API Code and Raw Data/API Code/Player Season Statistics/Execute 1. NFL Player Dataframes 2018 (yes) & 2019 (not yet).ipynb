```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

key = "e5fcs5mgz2fcmqpc3cmfmqcf"
```

```python

conf_df = []
div_df = []
ID_df = []
name_df = []
market_df = []
alias_df = []
url = "https://api.sportradar.us/nfl/official/trial/v5/en/league/hierarchy.json?api_key="+ key

###############################################                                       
#  add this back in when you need to call it  #
###############################################
JSONContent = requests.get(url).json()

conf_ct = len(JSONContent["conferences"])

for i in range (0, conf_ct):
    div_ct = len(JSONContent["conferences"][i])
    for j in range (0, div_ct): 
        team_ct = len(JSONContent["conferences"][i]["divisions"][j]["teams"])
        for k in range (0, team_ct):
            try: 
                #print(JSONContent)
                conf = JSONContent["conferences"][i]["name"]
                div = JSONContent["conferences"][i]["divisions"][j]["name"]
                vid = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["id"]
                name = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["name"]
                market = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["market"]
                alias = JSONContent["conferences"][i]["divisions"][j]["teams"][k]["alias"]
                conf_df.append(conf)
                div_df.append(div)
                ID_df.append(vid)
                name_df.append(name)
                market_df.append(market)
                alias_df.append(alias)
            except: 
                print("bye")

data = {"Conf": conf_df,"Division": div_df, "Team ID": ID_df, "Name": name_df, "Market": market_df, "Alias": alias_df}
team_df = pd.DataFrame(data) 
team_df
```
