```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import sys

key = "jxt3v9r9gdbvxhxcvarr7uc6"
```

```python
# File to Load 
team_data_to_load = "Team_List.csv"
players_2018_data_to_load = "Offense_Player_List_2018.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
players_2018_data = pd.read_csv(players_2018_data_to_load)
```

```python
#Conversions
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
offensive_pos = ["TE", "WR", "RB", "QB", "FB", "K"]
master_player_list = players_2018_data["Player ID"].tolist()
team_list = team_df["Team ID"].tolist()
teamname_list = team_df["Name"].tolist()


pid_df = []
season_df = []
ir_avg_yds_df = []
ir_longest_df =[] 
ir_longest_td_df =[] 
ir_returns_df =[] 
ir_touchdowns_df =[]
ir_yds_df =[]




for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url2 = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(1)
        JSONContent2 = requests.get(url2).json()
        player_ct = len(JSONContent2["players"])
        for j in range (0, player_ct):
            pid = JSONContent2["players"][j]["id"]
            if pid in master_player_list:
                try:
                    ir_avg_yds =JSONContent2["players"][j]["int_returns"]["avg_yards"] 
                    ir_longest =JSONContent2["players"][j]["int_returns"]["longest"] 
                    ir_longest_td =JSONContent2["players"][j]["int_returns"]["longest_touchdown"] 
                    ir_returns =JSONContent2["players"][j]["int_returns"]["returns"] 
                    ir_touchdowns =JSONContent2["players"][j]["int_returns"]["touchdowns"]
                    ir_yds =JSONContent2["players"][j]["int_returns"]["yards"]
                    season_df.append(yr)
                    pid_df.append(pid)
                    ir_avg_yds_df.append(ir_avg_yds)
                    ir_longest_df.append(ir_longest)
                    ir_longest_td_df.append(ir_longest_td)
                    ir_returns_df.append(ir_returns)
                    ir_touchdowns_df.append(ir_touchdowns)
                    ir_yds_df.append(ir_yds)


                except: 
                    print(sys.exc_info())  
                    continue 


data5 = {"Player Season": season_df,
        "Player ID": pid_df,
        "Int Ret Avg Yds":ir_avg_yds_df,
        "Int Ret Longest":ir_longest_df,
        "Int Ret Longest TD":ir_longest_td_df,
        "Int Ret Returns":ir_returns_df,
        "Int Ret Touchdowns":ir_touchdowns_df,
        "Int Ret Yds":ir_yds_df
      }

player_ir_df = pd.DataFrame(data5) 
player_ir_df.head()

```
