```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

key = "3tqkcnf7hs5wvwtryxyy2jru"
```

```python
# File to Load 
team_data_to_load = "Team_List.csv"
players_2018_data_to_load = "Offense_Player_List_2018.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
players_2018_data = pd.read_csv(players_2018_data_to_load)
```

```python
#fumbles
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
offensive_pos = ["TE", "WR", "RB", "QB", "FB", "K"]
master_player_list = players_2018_data["Player ID"].tolist()
team_list = team_df["Team ID"].tolist()
teamname_list = team_df["Name"].tolist()


fum_ez_rec_tds_df = []
fum_forced_df = []
fum_fum_df = []
fum_lost_fum_df = []
fum_opp_rec_df = []
fum_opp_rec_tds_df = []
fum_opp_rec_yds_df = []
fum_oob_df = []
fum_own_rec_df = []
fum_own_rec_tds_df = []
fum_own_rec_yds_df = []
pid_df = []
season_df = []

for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url6 = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(2)
        JSONContent6 = requests.get(url6).json()
        player_ct = len(JSONContent6["players"])
        for j in range (0, player_ct):
            pid = JSONContent6["players"][j]["id"]
            if pid in master_player_list:
                try:
                    fum_ez_rec_tds = JSONContent6["players"][j]["fumbles"]["ez_rec_tds"] 
                    fum_forced = JSONContent6["players"][j]["fumbles"]["forced_fumbles"]  
                    fum_fum = JSONContent6["players"][j]["fumbles"]["fumbles"] 
                    fum_lost_fum = JSONContent6["players"][j]["fumbles"]["lost_fumbles"] 
                    fum_opp_rec = JSONContent6["players"][j]["fumbles"]["opp_rec"] 
                    fum_opp_rec_tds = JSONContent6["players"][j]["fumbles"]["opp_rec_tds"] 
                    fum_opp_rec_yds = JSONContent6["players"][j]["fumbles"]["opp_rec_yards"] 
                    fum_oob = JSONContent6["players"][j]["fumbles"]["out_of_bounds"] 
                    fum_own_rec = JSONContent6["players"][j]["fumbles"]["own_rec"] 
                    fum_own_rec_tds = JSONContent6["players"][j]["fumbles"]["own_rec_tds"] 
                    fum_own_rec_yds = JSONContent6["players"][j]["fumbles"]["own_rec_yards"] 
                    fum_ez_rec_tds_df.append(fum_ez_rec_tds)
                    fum_forced_df.append(fum_forced)
                    fum_fum_df.append(fum_fum)
                    fum_lost_fum_df.append(fum_lost_fum)
                    fum_opp_rec_df.append(fum_opp_rec)
                    fum_opp_rec_tds_df.append(fum_opp_rec_tds)
                    fum_opp_rec_yds_df.append(fum_opp_rec_yds)
                    fum_oob_df.append(fum_oob)
                    fum_own_rec_df.append(fum_own_rec)
                    fum_own_rec_tds_df.append(fum_own_rec_tds)
                    fum_own_rec_yds_df.append(fum_own_rec_yds)
                    season_df.append(yr)
                    pid_df.append(pid)
                    print(pid)
                except:
                   continue


data5 = {"Player ID": pid_df,
        "Season": season_df,
        "Fumble End Zone Touch Downs": fum_ez_rec_tds_df,
        "Forced Fumbles": fum_forced_df,
        "Fumbles": fum_fum_df, 
        "Lost Fumbles": fum_lost_fum_df, 
        "Fumbles Opponent Recovery": fum_opp_rec_df,
        "Fumbles Opponent Recovery Touchdowns":fum_opp_rec_tds_df,
        "Fumbles Opponent Recovery Yards":fum_opp_rec_yds_df,
        "Fumbles Out of Bounds": fum_oob_df,
        "Fumbles Own Recovery":fum_own_rec_df,
        "Fumbles Own Recovery Touchdowns":fum_own_rec_tds_df,
        "Fumbles Own Recovery Yards":fum_own_rec_yds_df}

player_fum_df = pd.DataFrame(data5) 
player_fum_df.head()

    








```
