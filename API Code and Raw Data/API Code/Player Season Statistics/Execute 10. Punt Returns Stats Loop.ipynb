```python
import http.client
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import sys

key = "h53gmfp5pf94b9qwzavnfhwc"
```

```python
# File to Load 
team_data_to_load = "Team_List.csv"
players_2018_data_to_load = "Offense_Player_List_2018.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
players_2018_data = pd.read_csv(players_2018_data_to_load)
```

```python
#Rushing
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
offensive_pos = ["TE", "WR", "RB", "QB", "FB", "K"]
master_player_list = players_2018_data["Player ID"].tolist()
team_list = team_df["Team ID"].tolist()
teamname_list = team_df["Name"].tolist()


pid_df = []
season_df = []
pr_avg_yds_df = []
pr_fair_catches_df = []
pr_longest_df= []
pr_longest_td_df = []
pr_returns_df = []
pr_tds_df = []
pr_yds_df = []



for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url7 = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(1)
        JSONContent7 = requests.get(url7).json()
        player_ct = len(JSONContent7["players"])
        for j in range (0, player_ct):
            pid = JSONContent7["players"][j]["id"]
            if pid in master_player_list:
                try:
                    pr_avg_yds = JSONContent7["players"][j]["punt_returns"]["avg_yards"] 
                    pr_fair_catches = JSONContent7["players"][j]["punt_returns"]["faircatches"] 
                    pr_longest= JSONContent7["players"][j]["punt_returns"]["longest"] 
                    pr_longest_td = JSONContent7["players"][j]["punt_returns"]["longest_touchdown"] 
                    pr_returns = JSONContent7["players"][j]["punt_returns"]["returns"] 
                    pr_tds = JSONContent7["players"][j]["punt_returns"]["touchdowns"] 
                    pr_yds = JSONContent7["players"][j]["punt_returns"]["yards"] 
                    season_df.append(yr)
                    pid_df.append(pid)
                    pr_avg_yds_df.append(pr_avg_yds)
                    pr_fair_catches_df.append(pr_fair_catches)
                    pr_longest_df.append(pr_longest)
                    pr_longest_td_df.append(pr_longest_td)
                    pr_returns_df.append(pr_returns)
                    pr_tds_df.append(pr_tds)
                    pr_yds_df.append(pr_yds)
                    print(pid)
                except:
                   print(sys.exc_info())  
                   continue 


data5 = {"Player Season": season_df,
        "Player ID": pid_df,
        "Punt Returns Avg Yards":pr_avg_yds_df,
        "Punt Returns Fair Catches":pr_fair_catches_df,
        "Punt Returns Longest":pr_longest_df,
        "Punt Returns Longest TD":pr_longest_td_df,
        "Punt Returns Returns":pr_returns_df,
        "Punt Returns Touchdowns":pr_tds_df,
        "Punt Returns Yards":pr_yds_df
      }

player_pr_df = pd.DataFrame(data5) 
player_pr_df.head()

```
