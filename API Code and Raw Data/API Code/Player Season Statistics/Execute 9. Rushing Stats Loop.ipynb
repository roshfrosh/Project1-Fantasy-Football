```python
# File to Load 
team_data_to_load = "Team_List.csv"
players_2018_data_to_load = "Offense_Player_List_2018.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
players_2018_data = pd.read_csv(players_2018_data_to_load)
```

```python
#Rushing
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
offensive_pos = ["TE", "WR", "RB", "QB", "FB", "K"]
master_player_list = players_2018_data["Player ID"].tolist()
team_list = team_df["Team ID"].tolist()
teamname_list = team_df["Name"].tolist()


pid_df = []
season_df = []
rush_attempts_df= []
rush_avg_yards_df = []
rush_broken_tackles_df= [] 
rush_kneel_downs_df= [] 
rush_longest_df= []
rush_longest_td_df=[]
rush_rz_attempts_df=[]
rush_scrambles_df=[]
rush_tlost_df=[]
rush_tlost_yds_df=[]
rush_tds_df=[]
rush_yds_df=[]
rush_yds_after_cont_df=[]


for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url7 = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(1)
        JSONContent7 = requests.get(url7).json()
        player_ct = len(JSONContent7["players"])
        for j in range (0, player_ct):
            pid = JSONContent7["players"][j]["id"]
            if pid in master_player_list:
                try:
                    rush_attempts= JSONContent7["players"][j]["rushing"]["attempts"]  
                    rush_avg_yards = JSONContent7["players"][j]["rushing"]["avg_yards"]  
                    rush_broken_tackles= JSONContent7["players"][j]["rushing"]["broken_tackles"]  
                    rush_kneel_downs= JSONContent7["players"][j]["rushing"]["kneel_downs"]  
                    rush_longest= JSONContent7["players"][j]["rushing"]["longest"]  
                    rush_longest_td=JSONContent7["players"][j]["rushing"]["longest_touchdown"]  
                    rush_rz_attempts=JSONContent7["players"][j]["rushing"]["redzone_attempts"]  
                    rush_scrambles=JSONContent7["players"][j]["rushing"]["scrambles"]  
                    rush_tlost=JSONContent7["players"][j]["rushing"]["tlost"]  
                    rush_tlost_yds=JSONContent7["players"][j]["rushing"]["tlost_yards"]  
                    rush_tds=JSONContent7["players"][j]["rushing"]["touchdowns"]  
                    rush_yds=JSONContent7["players"][j]["rushing"]["yards"]  
                    rush_yds_after_cont=JSONContent7["players"][j]["rushing"]["yards_after_contact"]  
                    season_df.append(yr)
                    pid_df.append(pid)
                    rush_attempts_df.append(rush_attempts)
                    rush_avg_yards_df.append(rush_avg_yards)
                    rush_broken_tackles_df.append(rush_broken_tackles)
                    rush_kneel_downs_df.append(rush_kneel_downs)
                    rush_longest_df.append(rush_longest)
                    rush_longest_td_df.append(rush_longest_td)
                    rush_rz_attempts_df.append(rush_rz_attempts)
                    rush_scrambles_df.append(rush_scrambles)
                    rush_tlost_df.append(rush_tlost)
                    rush_tlost_yds_df.append(rush_tlost_yds)
                    rush_tds_df.append(rush_tds)
                    rush_yds_df.append(rush_yds)
                    rush_yds_after_cont_df.append(rush_yds_after_cont)
                    print(pid)
                except:
                   print(sys.exc_info()) 
                   continue 


data5 = {"Player Season": season_df,
        "Player ID": pid_df,
         "Rush Attempts": rush_attempts_df,
         "Rush Avg Yards":rush_avg_yards_df,
         "Rush Broken Tackles":rush_broken_tackles_df,
         "Rush Kneel Downs":rush_kneel_downs_df,
         "Rush Longest":rush_longest_df,
         "Rush Longest TD":rush_longest_td_df,
         "Rush Red Zone Attempts":rush_rz_attempts_df,
         "Rush Scrambles":rush_scrambles_df,
         "Rush Tackles for Loss":rush_tlost_df,
         "Rush Tackles for Loss Yards":rush_tlost_yds_df,
         "Rush Touchdowns":rush_tds_df,
         "Rush Yards":rush_yds_df,
         "Rush Yards After Contact":rush_yds_after_cont_df
      }

player_rush_df = pd.DataFrame(data5) 
player_rush_df.head()

```
